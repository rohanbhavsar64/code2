```
import pandas as pd
import plotly.graph_objects as go
```

```
df1=pd.read_excel('H1-H2-H3-H4-H5-HS_(21-11-2024)_COF.xlsx')
```

```
df1
```

```
fig = go.Figure(data=[
    go.Line(x=df1['Time(secs) - F 1'], y=df1['Coeff of Friction - F 1'],name='HH1'),
    go.Line(x=df1['Time(secs) - F 2'], y=df1['Coeff of Friction - F 2'],name='HH2'),
    go.Line(x=df1['Time(secs) - F 3'], y=df1['Coeff of Friction - F 3'],line_color='yellow',name='HH3'),
    go.Line(x=df1['Time(secs) - F 4'], y=df1['Coeff of Friction - F 4'],name='HH4'),
    go.Line(x=df1['Time(secs) - F 5'], y=df1['Coeff of Friction - F 5'],name='HH5'),
    go.Line(x=df1['Time(secs) - F 6'], y=df1['Coeff of Friction - F 6'],name='HHS'),

])
fig.update_layout(title='Coeff of Friction VS Time',
                  xaxis_title='Time',
                  yaxis_title='Coeff of Friction ')
```
